# Qwen2.5-VL-7B-Instruct vLLM Server

This notebook starts a vLLM server hosting the **Qwen2.5-VL-7B-Instruct** Vision Language Model with automatic tool-calling enabled.

## Model Features

- Multimodal (Text + Images)
- OpenAI-compatible API
- Automatic Tool Calling
- Function Calling Support
- Optimized Inference using vLLM

---

## Server Configuration

The server is launched with the following configuration:

```bash
vllm serve Qwen/Qwen2.5-VL-7B-Instruct \
  --host 0.0.0.0 \
  --port 8000 \
  --enable-auto-tool-choice \
  --tool-call-parser hermes \
  --max-model-len 4096 \
  --gpu-memory-utilization 0.4 \
  --chat-template ./hermes.jinja
```

### Configuration Details

| Parameter | Value | Description |
|------------|---------|-------------|
| Model | Qwen/Qwen2.5-VL-7B-Instruct | Vision Language Model |
| Host | 0.0.0.0 | Listen on all interfaces |
| Port | 8000 | API endpoint port |
| Auto Tool Choice | Enabled | Allows model to invoke tools automatically |
| Tool Parser | Hermes | Parses tool-calling responses |
| Context Length | 4096 | Maximum tokens per request |
| GPU Utilization | 40% | Limits GPU memory usage |
| Chat Template | hermes.jinja | Tool-calling compatible prompt template |

````python
import asyncio
import websockets
import cv2
import numpy as np
from IPython.display import display
from PIL import Image
import ipywidgets as widgets

img_widget = widgets.Image(format='jpeg')
display(img_widget)

async def stream_video():
    async with websockets.connect(
        "ws://localhost:8001/api/video/ws"
    ) as ws:

        while True:
            frame_bytes = await ws.recv()
            img_widget.value = frame_bytes

await stream_video()
````
